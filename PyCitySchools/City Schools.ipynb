```python
# Dependencies and Setup
import pandas as pd

# Load Files
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
# View new dataset
school_data_complete
```

```python
# Find total # of schools
unique_schools = school_data_complete["school_name"].unique()
total_schools = len(unique_schools)
print(total_schools)

# Find total # of students
total_students = school_data_complete["student_name"].count()
print(total_students)

# Find total budget - sum of budget for each school
unique_budget = school_data_complete["budget"].unique()
total_budget = sum(unique_budget)
print(total_budget)

# Find average math and reading scores
avg_math_score = school_data_complete["math_score"].mean()
print(avg_math_score)

avg_reading_score = school_data_complete["reading_score"].mean()
print(avg_reading_score)

# Calculate the % of students passing reading and math
math_passing = school_data_complete.loc[school_data_complete["math_score"] >= 70, :]
percent_pass_math = math_passing["student_name"].count() / school_data_complete["student_name"].count() * 100
print(percent_pass_math)

reading_passing = school_data_complete.loc[school_data_complete["reading_score"] >= 70, :]
percent_pass_reading = reading_passing["student_name"].count() / school_data_complete["student_name"].count() * 100
print(percent_pass_reading)

# Caclualte overall passing score
overall_passing = (avg_math_score + avg_reading_score)/2
overall_passing
```

```python
# Create dataframe to hold results
district_summary = pd.DataFrame({"Total Schools": total_schools},
                               {"Total Students": total_students},
                               {"Total Budget": total_budget},
                               {"Average Math Score": avg_math_score},
                               {"Average Reading Score": avg_reading_score},
                               {"% Passing Math": percent_pass_math},
                               {"% Passing Reading": percent_pass_reading},
                               {"% Overall Passing Rate": overall_passing})
district_summary
```
